#### Ejemplo 1

Escribir un algoritmo que permita calcular el cuadrado de los N primeros números enteros y a continuación escribir una tabla que contenga dichos cuadrados.

```python
""" Definicion de funciones """
def llenarVector(N):
    L = []
    for k in range(N):
        L += [k]
    return L

def elevarVector(V, pot):
    P = []
    for elemento in V:
        P += [elemento**pot]
    return P

def test_funciones():
    L1 = llenarVector(10)
    L2 = elevarVector(L1,3)
    print(L1)
    print(L2)

def main():
    num = int(input("Ingrese el numero final: "))
    V = llenarVector(num) # Tambien se pudo haber usado range: V = list(range(num))
    V_cuad = elevarVector(V,2)
    # Impresion de la tabla
    print("---------------------------------------------------------")
    print("Elementos de la lista - Elementos de la lista al cuadrado")
    print("---------------------------------------------------------")
    for i in range(len(V)):
        print(str(V[i]) + " --- " + str(V_cuad[i]))
    print("---------------------------------------------------------")

#test_funciones()  
""" Invocacion de funciones """
main()
```

#### Ejemplo 2

Disene la logica para un programa que permita a un usuario introducir 15 numeros y que luego los despliegue en el orden inverso al de entrada.

```python
""" Definicion de funciones """

# Funcion que invierta una lista
def invertirLista(L):
    L_inv = []
    for i in range(len(L)-1,-1,-1):
        L_inv += [L[i]]
    return L_inv

# Test de la funcion que invierte una lista
def test_invertirLista():
    L1 = [1, 2, 3, 4, 5, 6]
    L2 = invertirLista([1, 2, 3, 4, 5, 6])
    print(L1)
    print(L2)

# Funcion principal
def main():
    NUM_ELEMENTOS = 15
    V = []
    print("Ingrese los 15 elementos: ")
    for i in range(NUM_ELEMENTOS):
        e = int(input("V[" + str(i) + "] = "))
        V += [e]
    print("Vector ingresado = ", V, sep = "")
    # Invirtiendo elementos
    V_inv = invertirLista(V)
    print("Vector invertido = ", V_inv, sep = "")
    
                
""" Invocacion de funciones """
# test_invertirLista()
main()
```

#### Ejemplo 3

Modifique el programa del ejemplo anterior de modo que el usuario pueda introducir cualquier cantidad de numeros menores que 15 hasta que se introduzca un valor centinela.

```python
""" Definicion de funciones """

def invertirLista(L):
    L_inv = []
    for i in range(len(L)-1,-1,-1):
        L_inv += [L[i]]
    return L_inv


# Funcion principal
def main():
    V = []
    num = int(input("Ingrese un valor numerico menor que 15: "))
    while(num < 15):
        V += [num]
        num = int(input("Ingrese un valor numerico menor que 15: "))
        
    print("Vector ingresado = ", V, sep = "")
    # Invirtiendo elementos
    V_inv = invertirLista(V)
    print("Vector invertido = ", V_inv, sep = "")
    
""" Invocacion de funciones """
main()
```

#### Ejemplo 4

Diseñe una función que cuente el numero de palabras que existen en una cadena. Luego, escriba las respectivas funciones para llevar a cabo el test.

**Solución**: A continuación se muestra la jerarquia de las funciones implementadas.

![funciones](ejemplo_contar_palabras.png)

El codigo fuente se muesta a continuación

```python
""" Definicion de funciones """

# Funcion que quita espacios al principio 
def quitarEspaciosPrincipio(cadena):
    start = 0
    for l in cadena:
        if l == ' ':
            start += 1
        else:
            break        
    return cadena[start:]
    
# Funcion que quita espacios al final
def quitarEspaciosFinal(cadena):
    end = len(cadena)
    for i in range(len(cadena)-1,-1,-1):
        if cadena[i] != ' ':
            break
        else:
            end -= 1
    return cadena[:end]

# Funcion que quita espacios en los extremos (invoca las funciones que quitan espacios a cada lado)
def quitarEspaciosExtermos(cadena):
    cadena = quitarEspaciosPrincipio(cadena)
    cadena = quitarEspaciosFinal(cadena)
    return cadena

# Funcion que cuenta palabras
def contarPalabras(cadena):
    cadena = quitarEspaciosExtermos(cadena)
    palabras = 0
    if len(cadena) == 0:
        return palabras
    else:
        l1 = cadena[0]
        for i in range(1,len(cadena)):
            l2 = cadena[i]
            if l2 == ' ' and l2 != l1 or i == 1 and l2 != ' ':
                palabras += 1
            l1 = l2
        return palabras
                
# Test de las funciones que quitan espacios
def testQuitarEspacios():
    #cad1 = "Alo, soy Salomon el telefono savio del Colombiano"
    cad1 = "         Buenos dias     "
    cad2 = "      Julio Jaramillo"
    cad3 = "Oscar Agudelo     "
    cad4 = "Los Visconti"
    print(cad1,cad2,cad3,cad4,sep="\n")
    cad1 = quitarEspaciosExtermos(cad1)
    cad2 = quitarEspaciosExtermos(cad2)
    cad3 = quitarEspaciosExtermos(cad3)
    cad4 = quitarEspaciosExtermos(cad4)
    print()
    print(cad1,cad2,cad3,cad4,sep="\n")

# Test de las funciones que cuentan palabras
def testContarPalabras():
    cad1 = "Alo, soy Salomon el telefono savio del Colombiano"
    print(contarPalabras(cad1))
    print(contarPalabras(" Buenos dias pues"))
    print(contarPalabras("Julio      Jaramillo"))

print("--- Test quitar espacios ---")
testQuitarEspacios()
print()
print("--- Test contar palabras ---")
testContarPalabras()
```

#### Ejemplo 4

Adaptar el problema anterior para que pueda interactuar con el usuario.

```python
""" Definicion de funciones """

# Funcion que quita espacios al principio 
def quitarEspaciosPrincipio(cadena):
    start = 0
    for l in cadena:
        if l == ' ':
            start += 1
        else:
            break        
    return cadena[start:]
    
# Funcion que quita espacios al final
def quitarEspaciosFinal(cadena):
    end = len(cadena)
    for i in range(len(cadena)-1,-1,-1):
        if cadena[i] != ' ':
            break
        else:
            end -= 1
    return cadena[:end]

# Funcion que quita espacios en los extremos (invoca las funciones que quitan espacios a cada lado)
def quitarEspaciosExtermos(cadena):
    cadena = quitarEspaciosPrincipio(cadena)
    cadena = quitarEspaciosFinal(cadena)
    return cadena

# Funcion que cuenta palabras
def contarPalabras(cadena):
    cadena = quitarEspaciosExtermos(cadena)
    palabras = 0
    if len(cadena) == 0:
        return palabras
    else:
        l1 = cadena[0]
        for i in range(1,len(cadena)):
            l2 = cadena[i]
            if l2 == ' ' and l2 != l1 or i == 1 and l2 != ' ':
                palabras += 1
            l1 = l2
        return palabras

# Funcion principal (Interaccion con el usuario)
def main():
    print("--- Este programa cuenta el numero de palabras que tiene un frase --- ")
    while(1):
        frase = input("Digite una frase cualquiera: ")
        numPalabras = contarPalabras(frase)
        print("La frase ingresada tiene " + str(numPalabras))
        print()
        continuar = input("Desea ingresar otra frase? (y/n)")
        if continuar == 'n':
            break
    print("\nGracias por preferirnos")
    
""" Invocacion de funciones """
main()
```
